# Read the CSV and Perform Basic Data Cleaning

```python
# Import data
df = pd.read_csv("b_source_data/exoplanet_data.csv")
# print(df.info())

# Drop columns where all values are null
df = df.dropna(axis='columns', how='all')

# Drop rows containing null values
df = df.dropna()

# Display data info
print(df.info())
df.head()
```

# Pre-processing

Use `koi_disposition` for the y values

```python
# Split dataframe into X and y
X = df.drop("koi_disposition", axis=1)
y = df["koi_disposition"]
print(X.shape, y.shape)
```

```python
# Split X and y into training and testing groups
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42)
```

```python
# Display training data
X_train.head()
```

```python
# Scale the data with MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Determine Features of Influence - Random Forest Classifier Method

```python
# Create and score random forest classifier
rf_clf = RandomForestClassifier(n_estimators=1000)
rf_clf = rf_clf.fit(X_train, y_train)
print('Random Forest Score:')
rf_clf.score(X_train, y_train)
```

```python
# Sort the features by their importance
rf_feature_sort = sorted(zip(X.columns,rf_clf.feature_importances_),key=itemgetter(1), reverse=True)
# rf_feature_sort
```

```python
# Plot Random Forest Feature Importance
fig = plt.figure(figsize=[12,12])
plt.barh(*zip(* (rf_feature_sort)))
plt.xlabel('Feature Importance')
plt.ylabel('Feature Name')
plt.title('Random Forest Assessment')
plt.show()
```

# Choose Features of Importance
From the Random Forest Assessment plot you can see a knee in the feature importance between koi_duration_err1 and koi_prad. This indicates that features with importance exceeding 0.04 are good candidates for seeding predictive models

```python
# Select Random Forest Features of Interest
rf_features = [feature[0] for feature in rf_feature_sort if feature[1] > 0.04]
rf_features
```
